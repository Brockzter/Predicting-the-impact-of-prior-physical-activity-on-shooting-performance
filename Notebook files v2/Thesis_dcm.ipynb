```python
import pandas as pd
import math

#Width and height of the target 60x60cm and by assigning x and y with 30 we get the max dist of 60
#We will use these for both a complete miss and a hit on the target, but not inside the point rings.
MAX_DISTANCE = 30
```

```python
hit = pd.read_csv('Hit_v2.csv')
fire = pd.read_csv('Fire.csv')

hitX = hit['ImpactPosX'].tolist()
hitY = hit['ImpactPosY'].tolist()
plexi = hit['ColliderName'].tolist()
hit_or_not = fire['WasHit'].tolist()
bulletID_fire = fire['BulletId'].tolist()
bulletID_hit = hit['BulletId'].tolist()

#Account for the inversion of the SAVIT simulator and, since values recorded are in meters, multiply by 100.
#Account for the y-axis offset from the SAVIT simulator (1.3)
for ix in range(0, len(hitX)):
    hitX[ix] *= (-100)
    hitY[ix] += 1.3

    
#Unsure if this is achives anything.
i = 0
for ix in range(0, len(hit_or_not)-1):
    if hit_or_not[ix] == 0:
        i += 1
#print("Total missed shots?", i)

x = 0
plexi_index = []
missad_tavla_ix = []
#print("Length of X before:", len(hitX))

for ix in range(0, len(plexi)):
    if plexi[ix] == 'plexi' and plexi[ix + 1] == 'plexi':
        missad_tavla_ix.append(ix)
        hitX[ix]  = MAX_DISTANCE 
        hitY[ix]  = MAX_DISTANCE
        plexi[ix] = 9999
    if plexi[ix] == 'plexi':
        plexi_index.append(ix)
    
#print("Plexi length", len(plexi_index))
#print("hit x length", len(hitX))

for ix in range(0, len(plexi_index)):    
    if 'plexi' in plexi:
        del hitX[ix]
        del hitY[ix]
#print("Length  of X after removing 'plexi':", len(hitX))

hitXdf = pd.DataFrame(hitX)
hitYdf = pd.DataFrame(hitYdf)

p1 = get_dcm(5, 10)
p2 = get_dcm(35, 40)
p3 = get_dcm(65, 70)
p4 = get_dcm(95, 100)
p5 = get_dcm(125, 130)
p6 = get_dcm(155, 160)
p7 = get_dcm(185, 190)
p8 = get_dcm(215, 220)
p9 = get_dcm(245, 250)
p10 = get_dcm(275, 280)
p11 = get_dcm(305, 310)
p12 = get_dcm(310, 315)
p13 = get_dcm(340, 345)
p14 = get_dcm(370, 375)

print(p1)
print(p2)
print(p3)
print(p4)
print(p5)
print(p6)
print(p7)
print(p8)
print(p9)
print(p10)
print(p11)
print(p12)
print(p13)
print(p14)
```

```python
def get_hypo_in_range(a, b):
    xm = hitXdf.iloc[pd.np.r_[a:b]].mean()
    ym = hitYdf.iloc[pd.np.r_[a:b]].mean()
    return round(math.hypot(xm, ym), 3)

def get_dcm(a, b):
    dcm=[]
    for i in range(0, 5):
        dcm.append(get_hypo_in_range(a, b))
        a += 5
        b += 5
    return dcm
```
