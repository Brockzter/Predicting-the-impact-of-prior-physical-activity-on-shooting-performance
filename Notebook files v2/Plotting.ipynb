```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)
ax.plot([67, 70, 75, 79, 99, 107, 145, 147, 147, 149, 150, 159], 
        [7.69, 6.54, 5.95, 5.92, 6.25, 4.19, 4.05, 4.64, 7.02, 5.57, 5.27, 5.03])

plt.ylabel('DCM')
plt.xlabel('Heart rate')

x = [67, 70, 75, 79, 99, 107, 145, 147, 147, 149, 150, 159]
y = [7.69, 6.54, 5.95, 5.92, 6.25, 4.19, 4.05, 4.64, 7.02, 5.57, 5.27, 5.03]
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
plt.show()
```
